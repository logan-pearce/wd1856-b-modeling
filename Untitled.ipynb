```python
def CompositePTProfilePlot(z, output_file_name = 'PTprofiles'):    
    from bokeh.plotting import figure, show, output_file, save
    from bokeh.io import output_notebook
    from bokeh.models import LinearColorMapper, ColumnDataSource, LinearInterpolator
    from bokeh.models import  Range1d, LabelSet, Label, ColorBar, FixedTicker
    from bokeh.palettes import Magma256, Turbo256
    from bokeh.transform import linear_cmap
    output_notebook()

    tools = "hover, zoom_in, zoom_out, save, undo, redo, pan"
    tooltips = [
        ('filename', '@filename'),
    ]

    p = figure(width=800, height=850, y_axis_type="log", tools=tools, 
               tooltips=tooltips, toolbar_location="above")

    for i in range(len(z)):
        directory = z[i]
        #try:
        r = pd.read_csv(directory+'/cloud-free-atm-df.csv')
        pressure = r['pressure']
        temperature = r['temperature']
        datadf = pd.DataFrame(data={'plotx':temperature, 'ploty':pressure, 'filename':directory
                           })
        data=ColumnDataSource(data=datadf)
        p.line('plotx','ploty', source=data, alpha=0.2, color='red', line_width = 2)
#         except FileNotFoundError:
#             pass


    p.yaxis.axis_label = "Pressure [Bars]"
    p.xaxis.axis_label = "Temperature [K]"
    p.xaxis.axis_label_text_font_size = '20pt'
    p.yaxis.axis_label_text_font_size = '20pt'
    p.yaxis.major_label_text_font_size = "15pt"
    p.xaxis.major_label_text_font_size = "15pt"
    p.y_range.flipped = True
    output_file(output_file_name+".html")
    save(p)
    

CompositePTProfilePlot(z, output_file_name = 'PTProfiles-Initial-9-models')
```
